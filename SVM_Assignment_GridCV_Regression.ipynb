```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("insurance_pree.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indepen=dataset[['age', 'bmi', 'children', 'sex_male',
       'smoker_yes']]
dependen=dataset[['charges']]
```

```python
indepen
dependen
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param_grid={'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(indepen,dependen)
```

```python
re=grid.cv_results_
print("The best R2 value parameteres for SVM".format(grid.best_params_))
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("The Future prediction ={}".format(Future_Prediction))
```
