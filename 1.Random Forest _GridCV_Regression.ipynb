```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pree.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[["charges"]]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['mse','mae'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]}   
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The best parameters for RF prediction".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Enter the age: "))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("The Future Prediction for RT{}".format(Future_Prediction))
```
